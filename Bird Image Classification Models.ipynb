## Dataset
We use the **NABirds** dataset, which contains over 48,000 images of 555 North American bird species.
- Bounding boxes: Used to focus models on regions of interest.
- Train-test split: Based on train_test_split.txt provided in the dataset.
- Download Link: https://dl.allaboutbirds.org/nabirds

## Model Architectures Explored

We tested a variety of deep learning models, ensuring at least one per team member:
- ResNet-101
- DenseNet-201

Each model was trained on the same dataset split and with the same preprocessing pipeline for fair comparison.
